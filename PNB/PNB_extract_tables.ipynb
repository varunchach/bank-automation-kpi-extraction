# PNB — Extract KPIs from PDFs

Uses `PNBKPIExtractor` to extract KPIs from PNB_investor_PPT.pdf and PNB_CASA_Numbers_PPT.pdf.

**Input:** data/PNB_investor_PPT.pdf, data/PNB_CASA_Numbers_PPT.pdf

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd()))
sys.path.insert(0, str(Path.cwd() / "IB"))

from pnb_extractor import PNBKPIExtractor

DATA_DIR = Path("data")
investor_pdf = DATA_DIR / "PNB_investor_PPT.pdf"
casa_pdf = DATA_DIR / "PNB_CASA_Numbers_PPT.pdf"
```

```python
extractor = PNBKPIExtractor(investor_pdf, casa_pdf, data_dir=DATA_DIR)
kpis = extractor.extract()
df = extractor.to_dataframe()
display(df)
```
