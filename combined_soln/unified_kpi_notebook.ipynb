# Unified KPI — IB, PNB & HDFC Bank

Orchestrates individual IB, PNB, and HDFC Bank solutions, combines into a single DataFrame, and ranks KPIs (higher better for profits, lower better for costs).

```python
import sys
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT))
sys.path.insert(0, str(ROOT / "combined_soln"))
sys.path.insert(0, str(ROOT / "IB"))
sys.path.insert(0, str(ROOT / "PNB"))
sys.path.insert(0, str(ROOT / "HDFC"))

from IB.kpi_extractor import IndianBankKPIExtractor, ALL_KPI_KEYS
from PNB.pnb_extractor import PNBKPIExtractor
from HDFC.hdfc_extractor import HDFCKPIExtractor
from kpi_ranking import compute_rank, get_rank_direction
from kpi_report_format import build_report_df

DATA_DIR = ROOT / "data"
IB_INV = DATA_DIR / "IB_investor_PPT.pdf"
IB_CASA = DATA_DIR / "IB_CASA_Numbers_PPT.pdf"
PNB_INV = DATA_DIR / "PNB_investor_PPT.pdf"
PNB_CASA = DATA_DIR / "PNB_CASA_Numbers_PPT.pdf"
HDFC_INV = DATA_DIR / "HDFC_investor_PPT.pdf"
HDFC_CASA = DATA_DIR / "HDFC_Bank_CASA.pdf"
```

```python
print("Extracting Indian Bank...")
kpis_ib = IndianBankKPIExtractor(IB_INV, IB_CASA).extract()
print(f"  Done: {len([v for v in kpis_ib.values() if v is not None])}/46")

print("Extracting PNB...")
kpis_pnb = PNBKPIExtractor(PNB_INV, PNB_CASA, data_dir=DATA_DIR).extract()
print(f"  Done: {len([v for v in kpis_pnb.values() if v is not None])}/46")

print("Extracting HDFC Bank...")
kpis_hdfc = HDFCKPIExtractor(HDFC_INV, HDFC_CASA).extract()
print(f"  Done: {len([v for v in kpis_hdfc.values() if v is not None])}/46")
```

```python
df = build_report_df(kpis_ib, kpis_pnb, compute_rank, kpis_hdfc=kpis_hdfc)
display(df)
```
